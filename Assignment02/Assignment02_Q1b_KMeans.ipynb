# Question 1b: Segmentation using KMeans

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from skimage import io
```

```python
#Loading Images
dog = np.array(io.imread("/content/dog.jpeg"))
sunset = np.array(io.imread("/content/sunset.jpeg"))
books = np.array(io.imread("/content/books.jpeg"))
bird = np.array(io.imread("/content/bird.jpeg"))

fig,axes = plt.subplots(2, 2,figsize=(10, 5))
axes[0, 0].imshow(dog)
axes[0, 0].axis('off')
axes[0, 0].set_title("Dog")
axes[0, 1].imshow(sunset)
axes[0, 1].axis('off')
axes[0, 1].set_title("Sunset")
axes[1, 0].imshow(books)
axes[1, 0].axis('off')
axes[1, 0].set_title("Books")
axes[1, 1].imshow(bird)
axes[1, 1].axis('off')
axes[1, 1].set_title("Bird")

plt.show()
```

```python
def kmeans_segmentation(image):

  image_flatten = image.reshape(-1,3)
  resultant_images = []
  for cluster_value in range(3,7):
    kmeans = KMeans(n_clusters = cluster_value, n_init='auto')
    kmeans.fit(image_flatten)
    segmented_image = kmeans.cluster_centers_[kmeans.labels_]
    segmented_image = segmented_image.reshape(image.shape)
    segmented_image /= 255
    resultant_images.append(segmented_image)

  # Displaying Images

  fig,axes = plt.subplots(1,4,figsize = (20,5))

  for i in range(3,7):
    axes[i-3].imshow(resultant_images[i-3],aspect='auto')
    axes[i-3].axis('off')
    axes[i-3].set_title(f"Clusters = {i}")

  plt.show()
```

```python
kmeans_segmentation(dog)
```

```python
kmeans_segmentation(sunset)
```

```python
kmeans_segmentation(books)
```

```python
kmeans_segmentation(bird)
```
